```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

# 1. Load the Data
df = pd.read_csv(r"C:\Users\saikr\Downloads\Task 3 and 4_Loan_Data.csv")

# 2. Define our inputs (X) and the answer we want to predict (y)
features = ['credit_lines_outstanding', 'loan_amt_outstanding', 'total_debt_outstanding', 
            'income', 'years_employed', 'fico_score']

X = df[features]
y = df['default']

# 3. Train the Machine Learning Model (Logistic Regression)
model = LogisticRegression(max_iter=1000)
model.fit(X, y)

# 4. Create the Business Function for Charlie
def calculate_expected_loss(customer_data):
    """
    Takes in a dictionary of a customer's financial details.
    Returns the expected financial loss in dollars.
    """
    # Convert the single customer's data into a format the model can read
    customer_df = pd.DataFrame([customer_data])
    
    # Predict Probability of Default (PD)
    probability_of_default = model.predict_proba(customer_df)[0][1]
    
    # Calculate the Exposure at Default (the total loan amount)
    loan_amount = customer_data['loan_amt_outstanding']
    
    # Calculate Expected Loss (assuming 10% recovery rate, so the bank loses 90%)
    expected_loss = probability_of_default * 0.90 * loan_amount
    
    return expected_loss, probability_of_default

# Imagine a very risky customer: High debt, low income, terrible FICO score
risky_customer = {
    'credit_lines_outstanding': 5,
    'loan_amt_outstanding': 5000.00,
    'total_debt_outstanding': 15000.00,
    'income': 30000.00,
    'years_employed': 2,
    'fico_score': 500}

# Imagine a very safe customer: No debt, high income, great FICO score
safe_customer = {
    'credit_lines_outstanding': 0,
    'loan_amt_outstanding': 5000.00,
    'total_debt_outstanding': 5000.00,
    'income': 95000.00,
    'years_employed': 10,
    'fico_score': 750}

risky_loss, risky_pd = calculate_expected_loss(risky_customer)
safe_loss, safe_pd = calculate_expected_loss(safe_customer)

print(f"RISKY CUSTOMER -> Probability of Default: {risky_pd*100:.2f}% | Expected Loss: ${risky_loss:.2f}")
print(f"SAFE CUSTOMER  -> Probability of Default: {safe_pd*100:.2f}% | Expected Loss: ${safe_loss:.2f}")
```
